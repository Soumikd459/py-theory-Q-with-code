```python
def add(x, y):
    return x + y

def multi(x, y):
    return x * y

# Original behavior
print("Original add(9, 5):", add(9, 5))      # Output: 14
print("Original multi(9, 5):", multi(9, 5))  # Output: 45

# Monkey patching the functions
def patched_add(x, y):
    return x - y

def patched_multi(x, y):
    return x - y

# Apply monkey patching
add = patched_add
multi = patched_multi

# Testing the patched methods
print("Patched add(9, 5):", add(9, 5))      # Output: 4
print("Patched multi(9, 5):", multi(9, 5))  # Output: 4
```

```python
def second_largest(nums):
    if len(nums) < 2:
        return "List must have at least two elements"
    
    largest = float('-inf')
    second_largest = float('-inf')
    
    for num in nums:
        if num > largest:
            second_largest = largest
            largest = num
        elif num > second_largest and num != largest:
            second_largest = num

    return second_largest

# Example usage:
nums = [2, 5, 3, 8, 6, 10]
print("Second largest element:", second_largest(nums))

```

```python
class Animal:
    def __init__(self, name):
        self._name = name  # Encapsulated attribute
        
    def get_name(self):
        return self._name  # Getter method to access the name attribute
    
    def sound(self):
        pass  # Abstract method, to be implemented by subclasses


class Dog(Animal):
    def sound(self):
        return 'dog barks'  # Implementation of the sound method


class Cat(Animal):
    def sound(self):
        return 'cat meows'  # Implementation of the sound method


# Example usage
dog = Dog('Buddy')
cat = Cat('Whiskers')

# Accessing name attribute using getter method
print(dog.get_name())  # Output: Buddy
print(cat.get_name())  # Output: Whiskers

# Polymorphic behavior
print(dog.sound())  # Output: dog barks
print(cat.sound())  # Output: cat meows
```

```python
# Define a function that divides two numbers
def divide(x, y):
    try:
        result = x / y
        print("Division result:", result)
    except ZeroDivisionError:
        print("Error: Division by zero!")
    except TypeError:
        print("Error: Unsupported operand types!")
    except Exception as e:
        print("An error occurred:", e)

# Test the divide function with different inputs
divide(10, 5)   # Output: Division result: 5.0
divide(10, 0)   # Output: Error: Division by zero!
divide(10, '2') # Output: Error: Unsupported operand types!
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.amazon.com/s?i=specialty-aps&bbn=16225007011&rh=n%3A16225007011%2Cn%3A13896617011&language=en_US&currency=INR&ref=nav_em__nav_desktop_sa_intl_computers_tablets_0_2_6_4"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

response = requests.get(url, headers = headers)
soup = BeautifulSoup(response.content, 'html.parser')

product = []
price = []
rating = []
review = []

# Extract data from the HTML content
for product in soup.find_all("div", class_="a-section a-spacing-none"):# title-instructions-style
    product_name_tag = product.find("span", class_="a-size-mini")# a-size-medium
    product_price_tag = product.find("span", class_="a-price-whole") # a-price
    product_rating_tag = product.find("span", class_="a-row a-size-small")#  a-icon-alt
    product_review_count_tag = product.find("span", class_="a-size-base ")#  a-size-base
    
    
    if product_name_tag :
        product.append(product_name_tag.text.strip())
    else:
        product.append('unavailable')
        
    if product_price_tag:
        prices.append(product_price_tag.text.strip())
    else:
        prices.append("Not available")

    if product_rating_tag:
        ratings.append(product_rating_tag.text.strip())
    else:
        ratings.append("Not available")

    if product_review_count_tag:
        reviews.append(product_review_count_tag.text.strip())
    else:
        reviews.append("Not available")
        
        
data = {
    "Product": product,
    "Price": price,
    "Rating": rating,
    "Reviews": review
}
df = pd.DataFrame(data)

# Print the DataFrame
print(df)

```

```python
# Define a class representing a Car
class Car:
    def __init__(self, make, model, year):
        self._make = make  # Public attribute
        self._model = model  # Public attribute
        self._year = year  # Public attribute

    def get_make(self):
        return self._make

    def get_model(self):
        return self._model

    def get_year(self):
        return self._year

    def _calculate_age(self, current_year):
        return current_year - self._year

# Create an instance of Car
my_car = Car("Toyota", "Camry", 2018)

# Show the data (attributes) and functions
print("Data:")
print("Make:", my_car.get_make())
print("Model:", my_car.get_model())
print("Year:", my_car.get_year())

print("\nFunctions:")
print("Age:", my_car._calculate_age(2024))

# Bundle data (attributes) and methods (functions) into a single unit (a class)
class Car:
    def __init__(self, make, model, year):
        self._make = make  # Public attribute
        self._model = model  # Public attribute
        self._year = year  # Public attribute

    def get_make(self):
        return self._make

    def get_model(self):
        return self._model

    def get_year(self):
        return self._year

    def _calculate_age(self, current_year):
        return current_year - self._year

# Create an instance of Car
my_car = Car("Toyota", "Camry", 2018)

# Operate on the data (attributes) into a single unit (a class)
print("\nOperate on the data into a single unit (a class):")
print("Make:", my_car.get_make())
print("Model:", my_car.get_model())
print("Year:", my_car.get_year())
print("Age:", my_car._calculate_age(2024))

# Control access to that unit by calling the method to calculate the age
print("\nControl access to that unit:")
print("Age:", my_car._calculate_age(2024))
```

```python
def remove_consonant(string):
    vowels = {'a', 'e', 'i', 'o', 'u'}
    result = ''
    for char in string:
        if char in vowels:
            result += char
    return result

# Example usage
print(remove_consonant('soumik'))

```

```python
def remove_vowel(string):
    vowels = {'a', 'e', 'i', 'o', 'u'} #The characters in the set ('a', 'e', 'i', 'o', 'u') are lowercase,set handle both uppercase and lowercase vowels uniformly.
    result = ''                         # set allows for case-insensitive 
    for char in string:
        if char not in vowels:
            result += char
    return result

# Example usage
print(remove_vowel('soumik')) 
```
